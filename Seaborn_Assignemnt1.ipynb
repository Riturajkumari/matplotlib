#### Bar plot
A bar plot gives an estimate of the central tendency for a numeric variable with the height of each rectangle.

```python

# change the estimator from mean to
# standard deviation
sns.barplot(x ='sex', y ='total_bill', data = tips,palette ='plasma')
```

```python

# reading the dataset
tips = sns.load_dataset('tips')

sns.countplot(x ='sex', data = tips)
```

```python
# bor plot
# reading the dataset box plot
tips = sns.load_dataset('tips')

sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```
